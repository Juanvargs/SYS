## Análisis de sistema masa-resorte-amortiguador

### Ecuaciones del sistema

$$
\sum F_y = ma
$$

- $$ F_s \rightarrow \text{Fuerza del resorte} = ky(t) $$
- $$ F_f \rightarrow \text{Fuerza de fricción} = c\dot{y}(t) $$
- $$ ma \rightarrow \text{Fuerza inercial} = m\ddot{y}(t) $$
- $$ f_E \rightarrow \text{Fuerza externa} = f_E(t) $$

$$
f_E - F_s - F_f = ma
$$

$$
f_E = F_s + F_f + ma
$$

$$
f_E = m\ddot{y}(t) + c\dot{y}(t) + ky(t)
$$

### Aplicamos Transformada de Laplace

$$
F_E(s) = m s^2 Y(s) + c s Y(s) + k Y(s)
$$

### Función de transferencia

$$
G(s) = \frac{Y(s)}{F_E(s)}
$$

### Aplico Factor común Y(s)

$$
F_E(s) = Y(s)(ms^2 + cs + k)
$$

### Entonces:

$$
G(s) = \frac{1}{ms^2 + cs + k}
$$


## Comparación Red Eléctrica — Sistema Mecánico

La ecuación del circuito es la siguiente:

**Circuito de segundo orden** porque tiene **dos elementos almacenadores de energía**:

$$
V_{in}(t) = V_{L}(t) + V_{o}(t)
$$

Sabemos que:

$$
V_L(t) = L \frac{di}{dt} + V_o(t) \tag{1}
$$

Corriente que llega por la malla:

$$
i(t) = i_C(t) + i_R(t)
$$

Donde:

$$
i_C(t) = C \frac{dV_o}{dt}
$$

y

$$
i_R(t) = \frac{V_o(t)}{R}
$$

Entonces:

$$
i(t) = C \frac{dV_o}{dt} + \frac{V_o(t)}{R} \tag{2}
$$

Derivando (2) respecto a t:

$$
\frac{di}{dt} = C \frac{d^2V_o}{dt^2} + \frac{1}{R} \frac{dV_o}{dt}
$$

Sustituyendo $\frac{di}{dt}$ en (1):

$$
V_{in}(t) = L \left( C \frac{d^2V_o}{dt^2} + \frac{1}{R} \frac{dV_o}{dt} \right) + V_o(t)
$$

## Función de Transferencia (condiciones iniciales cero)

Aplicando Transformada de Laplace:

$$
L \left( C s^2 V_o(s) + \frac{1}{R} s V_o(s) \right) + V_o(s) = V_{in}(s)
$$

Factorizando:

$$
\left( L C s^2 + \frac{L}{R} s + 1 \right) V_o(s) = V_{in}(s)
$$

**Función de transferencia**:

$$
G(s) = \frac{V_o(s)}{V_{in}(s)} = \frac{1}{L C s^2 + \frac{L}{R} s + 1}
$$

## Comparación con Sistema Masa-Resorte-Amortiguador:

Forma general:

$$
G(s) = \frac{1}{m s^2 + c s + k}
$$

### Correspondencia:

| Sistema Mecánico | Sistema Eléctrico |
|::----------------|:------------------|
| $m$              | $L C$              |
| $c$              | $\frac{L}{R}$      |
| $k$              | $1$                |

# 📖 Modelo Matemático de la Modulación y Demodulación SSB-AM

---

## 📌 Señales Involucradas

- Señal de mensaje:
  
$$m(t)$$

- Frecuencia de la portadora:

$$f_c$$

- Señal portadora:

$$c(t) = \cos(2 \pi f_c t)$$

---

## 📌 1️⃣ Modulación SSB-AM

### 📌 📎 Dominio del Tiempo

La señal modulada en banda lateral única (SSB-AM) se expresa como:

$$
s_{SSB}(t) = m(t) \cos(2 \pi f_c t) \pm \hat{m}(t) \sin(2 \pi f_c t)
$$

Donde:
- $$\hat{m}(t)$$ es la **transformada de Hilbert** de $$m(t)$$
- El signo $$+$$ corresponde a la **Banda Lateral Superior (USB)**
- El signo $$-$$ corresponde a la **Banda Lateral Inferior (LSB)**

**Transformada de Hilbert:**

$$
\hat{m}(t) = \frac{1}{\pi t} * m(t)
$$

---

### 📌 📎 Dominio de la Frecuencia (vía Transformada de Fourier)

Sea $$M(f)$$ la Transformada de Fourier de $$m(t)$$

**Transformada de Hilbert en frecuencia:**

$$
\mathcal{F}\{\hat{m}(t)\} = -j \cdot \text{sgn}(f) \cdot M(f)
$$

La Transformada de Fourier de:

$$
s_{SSB}(t) = m(t) \cos(2 \pi f_c t) \pm \hat{m}(t) \sin(2 \pi f_c t)
$$

es:

$$
S_{SSB}(f) = \frac{1}{2} \left[ M(f - f_c) \pm j \cdot \text{sgn}(f - f_c) \cdot M(f - f_c) + M(f + f_c) \mp j \cdot \text{sgn}(f + f_c) \cdot M(f + f_c) \right]
$$

Para conservar **solo la USB**:

$$
S_{USB}(f) =
\begin{cases}
M(f - f_c), & f > 0 \\
0, & f < 0
\end{cases}
$$

Y para conservar **solo la LSB**:

$$
S_{LSB}(f) =
\begin{cases}
0, & f > 0 \\
M(f + f_c), & f < 0
\end{cases}
$$

---

## 📌 2️⃣ Demodulación SSB-AM

### 📌 📎 Dominio del Tiempo

Se multiplica la señal $$s_{SSB}(t)$$ por una copia de la portadora:

$$
y(t) = s_{SSB}(t) \cdot \cos(2 \pi f_c t)
$$

Sustituyendo:

$$
= \left[m(t) \cos(2 \pi f_c t) \pm \hat{m}(t) \sin(2 \pi f_c t)\right] \cdot \cos(2 \pi f_c t)
$$

**Usando identidades trigonométricas:**

$$
\cos^2(\theta) = \frac{1 + \cos(2 \theta)}{2}
$$

$$
\sin(\theta)\cos(\theta) = \frac{\sin(2 \theta)}{2}
$$

Entonces:

$$
y(t) = \frac{m(t)}{2} [1 + \cos(4 \pi f_c t)] \pm \frac{\hat{m}(t)}{2} \sin(4 \pi f_c t)
$$

Se pasa por un **filtro pasa-bajas**, eliminando los términos en $$2 f_c$$, y queda:

$$
y(t) = \frac{m(t)}{2}
$$

---

### 📌 📎 Dominio de la Frecuencia (vía Transformada de Fourier)

La Transformada de Fourier de la señal demodulada es:

$$
Y(f) = \frac{1}{2} \left[ S_{SSB}(f - f_c) + S_{SSB}(f + f_c) \right]
$$

Luego, al aplicar un **filtro pasa-bajas (PB)** de ancho $$|f| \leq B$$:

$$
Y(f) =
\begin{cases}
\frac{1}{2} M(f), & |f| \leq B \\
0, & \text{en otro caso}
\end{cases}
$$

---

## 📌 📖 Propiedades Utilizadas

**Transformada de Fourier de un coseno:**

$$
\mathcal{F}\{\cos(2 \pi f_c t)\} = \frac{1}{2} \left[ \delta(f - f_c) + \delta(f + f_c) \right]
$$

**Transformada de Hilbert en frecuencia:**

$$
\mathcal{F}\{\hat{m}(t)\} = -j \cdot \text{sgn}(f) \cdot M(f)
$$

**Convolución en frecuencia:**

$$
\mathcal{F}\{x(t) \cdot y(t)\} = X(f) * Y(f)
$$

```python
%%writefile pages/1_⚙️_Simulador_de_Sistemas.py

import streamlit as st
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# --- Configuración de la página ---
st.set_page_config(
    page_title="Simulación de Sistemas de Segundo Orden",
    page_icon="⚙️",
    layout="wide",
)

st.markdown("""
    <h1 style='color:#33e6ff; font-size: 36px;'>⚙️ Simulador de Sistemas de Segundo Orden</h1>
    <p style='font-size:18px;'>Explora el comportamiento dinámico de sistemas masa-resorte-amortiguador y su circuito RLC equivalente.</p>
""", unsafe_allow_html=True)

# --- Barra lateral ---
st.sidebar.title("🧩 Configuración del Sistema")

response_type = st.sidebar.selectbox(
    "Tipo de Sistema",
    ("Subamortiguada", "Críticamente Amortiguada", "Sobreamortiguada", "Inestable"),
)

if response_type == "Subamortiguada":
    zeta = st.sidebar.slider("Amortiguamiento (ζ)", 0.01, 0.99, 0.3, 0.01)
elif response_type == "Críticamente Amortiguada":
    zeta = 1.0
    st.sidebar.info("ζ = 1 para amortiguamiento crítico")
elif response_type == "Sobreamortiguada":
    zeta = st.sidebar.slider("Amortiguamiento (ζ)", 1.1, 5.0, 1.5, 0.1)
else:
    zeta = st.sidebar.slider("Amortiguamiento (ζ)", -1.0, -0.01, -0.5, 0.01)

omega_n = st.sidebar.slider("Frecuencia Natural (ωₙ) [rad/s]", 1.0, 20.0, 5.0, 0.5)


# --- Definición de sistemas ---
num_ol = [omega_n ** 2]
den_ol = [1, 2 * zeta * omega_n, omega_n ** 2]
sys_ol = signal.TransferFunction(num_ol, den_ol)

den_cl = [den_ol[0], den_ol[1], den_ol[2] + num_ol[0]]
sys_cl = signal.TransferFunction(num_ol, den_cl)

# --- Componentes físicos ---
m = 1.0
c = 2 * zeta * omega_n * m
k = omega_n ** 2 * m
C_elec = 1.0
if zeta != 0 and omega_n != 0:
    R_elec = 1 / (2 * zeta * omega_n * C_elec)
    L_elec = 1 / ((omega_n ** 2) * C_elec)
else:
    R_elec, L_elec = float('inf'), float('inf')

# --- Parámetros temporales ---
ts, tp, mp, tr = None, None, None, None
if 0 < zeta < 1:
    ts = 4 / (zeta * omega_n)
    tp = np.pi / (omega_n * np.sqrt(1 - zeta ** 2))
    mp = 100 * np.exp((-zeta * np.pi) / np.sqrt(1 - zeta ** 2))
    t_step_calc, y_step_calc = signal.step(sys_cl)
    try:
        final_value = y_step_calc[-1]
        tr_10 = t_step_calc[np.where(y_step_calc >= 0.1 * final_value)[0][0]]
        tr_90 = t_step_calc[np.where(y_step_calc >= 0.9 * final_value)[0][0]]
        tr = tr_90 - tr_10
    except IndexError:
        tr = "No calculable"
elif zeta < 0:
    ts = tr = mp = tp = "No aplica"
else:
    mp = 0.0
    tp = "No aplica"
    t_step_calc, y_step_calc = signal.step(sys_cl)
    final_value = y_step_calc[-1]
    if final_value > 1e-6:
        try:
            settling_indices = np.where(np.abs(y_step_calc - final_value) >= 0.02 * final_value)[0]
            ts = t_step_calc[settling_indices[-1]] if settling_indices.size > 0 else 0.0
            tr_10_idx = np.where(y_step_calc >= 0.1 * final_value)[0]
            tr_90_idx = np.where(y_step_calc >= 0.9 * final_value)[0]
            tr = t_step_calc[tr_90_idx[0]] - t_step_calc[tr_10_idx[0]] if tr_10_idx.size > 0 and tr_90_idx.size > 0 else "No calculable"
        except:
            ts, tr = "Error", "Error"
    else:
        ts, tr = "N/A", "N/A"

# --- Visualización ---
col1, col2 = st.columns(2)

with col1:
    st.subheader("📌 Parámetros del Sistema")
    st.markdown(f"**Sistema Mecánico:**\n- Masa (m): {m:.2f} kg\n- Constante (k): {k:.2f} N/m\n- Amortiguador (c): {c:.2f} Ns/m")
    st.markdown(f"**Circuito RLC equivalente:**\n- R = {R_elec:.2f} Ω\n- L = {L_elec:.2f} H\n- C = {C_elec:.2f} F")

with col2:
    st.markdown("<h3 style='color: #33e6ff;'>Parámetros Temporales (Lazo Cerrado)</h3>", unsafe_allow_html=True)
    st.markdown("A continuación se detallan los parámetros que caracterizan la respuesta temporal del sistema ante una entrada escalón unitario.")

    if 0 < zeta < 1:
        # Subamortiguado
        st.markdown("**1. Tiempo de levantamiento (Tr):** Tiempo que tarda la salida en subir del 10% al 90% del valor final.")
        st.code(f"{tr:.3f} s", language="text")

        st.markdown("**2. Sobreimpulso máximo (Mp):** Porcentaje que la salida sobrepasa el valor final antes de estabilizarse.")
        st.code(f"{mp:.2f} %", language="text")

        st.markdown("**3. Tiempo de pico (Tp):** Tiempo en el que se alcanza el máximo valor de la respuesta.")
        st.code(f"{tp:.3f} s", language="text")

        st.markdown("**4. Tiempo de establecimiento (Ts):** Tiempo que tarda la respuesta en mantenerse dentro del ±2% del valor final.")
        st.code(f"{ts:.3f} s", language="text")

    elif zeta < 0:
        # Inestable
        st.warning("⚠️ Para un sistema inestable, los parámetros temporales no están definidos.")
    else:
        # Crítico o sobreamortiguado
        st.info("ℹ️ Para sistemas sobreamortiguados o críticamente amortiguados, no hay sobreimpulso ni tiempo de pico.")

        st.markdown("**1. Tiempo de establecimiento (Ts):** Tiempo en que la respuesta entra y permanece dentro del ±2% del valor final.")
        st.code(f"{ts if isinstance(ts, str) else f'{ts:.3f} s'}", language="text")

        st.markdown("**2. Tiempo de levantamiento (Tr):** Tiempo en que la respuesta pasa del 10% al 90% del valor final.")
        st.code(f"{tr if isinstance(tr, str) else f'{tr:.3f} s'}", language="text")

st.divider()

# --- Tabs ---
tab1, tab2, tab3, tab4, tab5 = st.tabs(["📉 Bode", "📍 Polos/Ceros", "⚡ Impulso", "⏱️ Escalón", "↗️ Rampa"])

def plot_pz(system, ax, title):
    ax.scatter(np.real(system.poles), np.imag(system.poles), marker='x', color='crimson', label='Polos')
    if system.zeros.size:
        ax.scatter(np.real(system.zeros), np.imag(system.zeros), marker='o', facecolors='none', edgecolors='navy', label='Ceros')
    ax.axhline(0, color='gray', lw=0.5)
    ax.axvline(0, color='gray', lw=0.5)
    ax.set_title(title)
    ax.grid(True)
    ax.legend()

with tab1:
    fig, (ax1, ax2) = plt.subplots(2,1, figsize=(10,6))
    w1, mag1, phase1 = signal.bode(sys_ol)
    w2, mag2, phase2 = signal.bode(sys_cl)
    ax1.semilogx(w1, mag1, label='Lazo Abierto')
    ax1.semilogx(w2, mag2, linestyle='--', label='Lazo Cerrado')
    ax1.set_ylabel("Magnitud (dB)")
    ax1.grid(True)
    ax1.legend()
    ax2.semilogx(w1, phase1, label='Lazo Abierto')
    ax2.semilogx(w2, phase2, linestyle='--', label='Lazo Cerrado')
    ax2.set_ylabel("Fase (°)")
    ax2.set_xlabel("Frecuencia (rad/s)")
    ax2.grid(True)
    ax2.legend()
    st.pyplot(fig)

with tab2:
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4))
    plot_pz(sys_ol, ax1, "Lazo Abierto")
    plot_pz(sys_cl, ax2, "Lazo Cerrado")
    st.pyplot(fig)

with tab3:
    t, y1 = signal.impulse(sys_ol)
    _, y2 = signal.impulse(sys_cl, T=t)
    fig, ax = plt.subplots()
    ax.plot(t, y1, label='Abierto')
    ax.plot(t, y2, linestyle='--', label='Cerrado')
    ax.set_title("Respuesta al Impulso")
    ax.set_xlabel("Tiempo (s)")
    ax.grid(True)
    ax.legend()
    st.pyplot(fig)

with tab4:
    t, y1 = signal.step(sys_ol)
    _, y2 = signal.step(sys_cl, T=t)
    fig, ax = plt.subplots()
    ax.plot(t, y1, label='Abierto')
    ax.plot(t, y2, linestyle='--', label='Cerrado')
    ax.axhline(1, color='gray', linestyle=':')
    ax.set_title("Respuesta al Escalón")
    ax.set_xlabel("Tiempo (s)")
    ax.grid(True)
    ax.legend()
    st.pyplot(fig)

with tab5:
    sys_ol_ramp = signal.TransferFunction(sys_ol.num, np.polymul(sys_ol.den, [1,0]))
    sys_cl_ramp = signal.TransferFunction(sys_cl.num, np.polymul(sys_cl.den, [1,0]))
    t, y1 = signal.step(sys_ol_ramp)
    _, y2 = signal.step(sys_cl_ramp, T=t)
    fig, ax = plt.subplots()
    ax.plot(t, t, 'k:', label='Entrada Rampa')
    ax.plot(t, y1, label='Abierto')
    ax.plot(t, y2, linestyle='--', label='Cerrado')
    ax.set_title("Respuesta a la Rampa")
    ax.set_xlabel("Tiempo (s)")
    ax.grid(True)
    ax.legend()
    st.pyplot(fig)
```

```python
%%writefile pages/2_⚡_SSB_AM.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, sosfilt, sosfreqz, tf2zpk
from scipy.fft import fft, ifft, fftfreq, fftshift, ifftshift
from scipy.io import wavfile
import os
import subprocess
import yt_dlp

st.set_page_config(page_title="Modulación SSB-AM", page_icon="📡", layout="wide")
st.title("📡 Modulación y Demodulación SSB-AM con Filtros IIR")

# ------------------ FUNCIONES AUXILIARES ------------------

def extraer_polos_y_zeros_validos(sos):
    """Extrae polos y ceros válidos desde una estructura SOS (ignora secciones inválidas)."""
    zeros, poles = [], []
    for section in sos:
        b, a = section[:3], section[3:]
        if np.allclose(a, 0):
            continue
        try:
            z, p, _ = tf2zpk(b, a)
            zeros.extend(z)
            poles.extend(p)
        except Exception:
            continue
    return np.array(zeros), np.array(poles)

def plot_bode_y_pz(sos, fs):
    fig, axs = plt.subplots(1, 2, figsize=(12, 4))

    # Diagrama de Bode
    w, h = sosfreqz(sos, worN=2000, fs=fs)
    axs[0].plot(w, 20 * np.log10(np.abs(h)))
    axs[0].set_title("Diagrama de Bode")
    axs[0].set_xlabel("Frecuencia (Hz)")
    axs[0].set_ylabel("Magnitud (dB)")
    axs[0].grid(True)

    # Plano de polos y ceros
    z, p = extraer_polos_y_zeros_validos(sos)
    axs[1].scatter(np.real(z), np.imag(z), marker='o', facecolors='none', edgecolors='b', label='Ceros')
    axs[1].scatter(np.real(p), np.imag(p), marker='x', color='r', label='Polos')
    axs[1].axhline(0, color='black', lw=0.5)
    axs[1].axvline(0, color='black', lw=0.5)
    axs[1].set_title("Plano de Polos y Ceros")
    axs[1].grid(True)
    axs[1].legend()

    return fig

def load_audio_from_youtube(url, duration_s=5, start_s=0):
    temp_wav_file = 'temp_audio.wav'
    cropped_wav_file = 'cropped_audio.wav'

    try:
        ydl_opts = {
            'format': 'bestaudio/best',
            'postprocessors': [{'key': 'FFmpegExtractAudio', 'preferredcodec': 'wav'}],
            'outtmpl': 'temp_audio',
            'quiet': True,
        }
        if os.path.exists(temp_wav_file):
            os.remove(temp_wav_file)
        if os.path.exists(cropped_wav_file):
            os.remove(cropped_wav_file)

        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])

        command = [
            'ffmpeg', '-i', temp_wav_file, '-ss', str(start_s),
            '-t', str(duration_s), '-acodec', 'pcm_s16le',
            '-ar', '44100', '-ac', '1', cropped_wav_file, '-y'
        ]
        subprocess.run(command, check=True, capture_output=True)

        fs_audio, audio_data = wavfile.read(cropped_wav_file)
        if audio_data.ndim > 1:
            audio_data = audio_data.mean(axis=1)
        audio_data = audio_data / np.max(np.abs(audio_data))
        m_t = audio_data
        t = np.linspace(start_s, start_s + len(m_t) / fs_audio, len(m_t), endpoint=False)

        return m_t, fs_audio, t
    except Exception as e:
        st.error(f"Error al procesar el audio de YouTube: {e}")
        return None, None, None

def ssb_modulate(m_t, t, fs, fc, tipo='USB'):
    from scipy.signal import hilbert
    m_hilbert = np.imag(hilbert(m_t))
    if tipo == 'USB':
        s_t = m_t * np.cos(2 * np.pi * fc * t) - m_hilbert * np.sin(2 * np.pi * fc * t)
    else:
        s_t = m_t * np.cos(2 * np.pi * fc * t) + m_hilbert * np.sin(2 * np.pi * fc * t)
    return s_t

def ssb_demodulate(s_t, t, fs, fc, sos_pb):
    demod = s_t * 2 * np.cos(2 * np.pi * fc * t)
    demod_filtered = sosfilt(sos_pb, demod)
    return demod_filtered

def plot_time(t, sig, title):
    fig, ax = plt.subplots(figsize=(10, 3))
    ax.plot(t, sig)
    ax.set_title(title)
    ax.set_xlabel("Tiempo (s)")
    ax.set_ylabel("Amplitud")
    ax.grid(True)
    return fig

def plot_freq(sig, fs, title, xlim=None):
    N = len(sig)
    f = fftfreq(N, 1 / fs)
    Y = np.abs(fft(sig))
    fig, ax = plt.subplots(figsize=(10, 3))
    ax.plot(fftshift(f), fftshift(Y))
    ax.set_title(title)
    ax.set_xlabel("Frecuencia (Hz)")
    ax.set_ylabel("Magnitud")
    if xlim: ax.set_xlim(xlim)
    ax.grid(True)
    return fig

# ------------------ INTERFAZ ------------------

st.sidebar.title("⚙️ Parámetros")
tipo_mensaje = st.sidebar.selectbox("Tipo de señal mensaje", ["Pulso Rectangular", "Audio de YouTube"])
fc = st.sidebar.slider("Frecuencia de portadora (Hz)", 1000, 20000, 10000, step=500)
lado = st.sidebar.radio("Seleccione la banda lateral", ["USB", "LSB"])
duracion = st.sidebar.slider("Duración (s)", 1, 10, 5)
fs = 44100

if tipo_mensaje == "Pulso Rectangular":
    t = np.linspace(0, duracion, duracion * fs, endpoint=False)
    ancho_pulso = st.sidebar.slider("Ancho del pulso (s)", 0.1, 1.0, 0.5)
    m_t = np.zeros_like(t)
    m_t[t < ancho_pulso] = 1
else:
    url = st.sidebar.text_input("URL de YouTube", "https://www.youtube.com/watch?v=QH2-TGUlwu4")
    inicio = st.sidebar.slider("Inicio (s)", 0, 60, 5)
    m_t, fs, t = load_audio_from_youtube(url, duration_s=duracion, start_s=inicio)

if m_t is not None:
    st.subheader("🔊 Señal de Mensaje")
    st.pyplot(plot_time(t, m_t, "Mensaje en Tiempo"))
    st.pyplot(plot_freq(m_t, fs, "Mensaje en Frecuencia", xlim=(-10000, 10000)))
    if tipo_mensaje == "Audio de YouTube":
        st.audio(m_t, format="audio/wav", sample_rate=fs)

    st.subheader("📡 Señal Modulada SSB")
    s_t = ssb_modulate(m_t, t, fs, fc, tipo=lado)
    st.pyplot(plot_time(t, s_t, "Señal SSB en Tiempo"))
    st.pyplot(plot_freq(s_t, fs, "Señal SSB en Frecuencia", xlim=(-20000, 20000)))

    st.subheader("🔁 Señal Demodulada")
    sos_pb = butter(10, 4000, fs=fs, output='sos')
    m_rec = ssb_demodulate(s_t, t, fs, fc, sos_pb)
    st.pyplot(plot_time(t, m_rec, "Señal Recuperada en Tiempo"))
    st.pyplot(plot_freq(m_rec, fs, "Señal Recuperada en Frecuencia", xlim=(-10000, 10000)))
    if tipo_mensaje == "Audio de YouTube":
        st.audio(m_rec, format="audio/wav", sample_rate=fs)

    st.subheader("🧭 Análisis del Filtro IIR Pasa Bajas")
    st.pyplot(plot_bode_y_pz(sos_pb, fs))
else:
    st.warning("Por favor selecciona o espera la carga de una señal de mensaje.")
```
